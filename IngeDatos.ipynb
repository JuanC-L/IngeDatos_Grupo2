```python
resultados = pd.DataFrame(columns=['Columna', 'Valores Únicos','Total', 'Valores NaN'])

for columna in data.columns:
    valores_unicos = len(data[columna].unique())
    total = data[columna].count()
    valores_nan = data[columna].isna().sum()
    resultados = resultados.append({'Columna': columna, 'Valores Únicos': valores_unicos,'Total':total,'Valores NaN': valores_nan}, ignore_index=True)

resultados.sort_values(by='Valores Únicos')
```

```python
books=pd.read_csv('books.csv')

books = books.drop_duplicates(subset='title', keep='first')

data4['ISBN10'].fillna(data4['title'].map(books.set_index('title')['isbn']), inplace=True)
```

```python

data4['item_weight_numeric'] = data4['item_weight'].str.extract('(\d+\.\d+|\d+)').astype(float)

data4['item_weight_text'] = data4['item_weight'].str.extract('([a-zA-Z]+)')

unit_conversion = {'ounces': 1/16, 'pounds': 1, 'libs': 1}

data4['item_weight_pounds'] = data4['item_weight_text'].map(unit_conversion) * data4['item_weight_numeric']

```

```python
data4['item_weight_pounds'].fillna(data4.item_weight_pounds.median(),inplace=True)
data4 = data4.drop(columns =(['item_weight_text','item_weight_numeric','item_weight']))
```

```python
resultados = pd.DataFrame(columns=['Columna', 'Valores Únicos', 'Valores NaN'])

for columna in data4.columns:
    valores_unicos = len(data4[columna].unique())
    total = data4[columna].count()
    valores_nan = data4[columna].isna().sum()
    resultados = resultados.append({'Columna': columna, 'Valores Únicos': valores_unicos,'Total':total,'Valores NaN': valores_nan}, ignore_index=True)

resultados.sort_values(by='Valores NaN', ascending=False)
```
